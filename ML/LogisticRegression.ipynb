```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
n = 200


age = np.random.randint(18, 60, size=n)
salary = np.random.randint(20000, 100000, size=n)

# Simulate 'Buy' based on some logic (higher age + higher salary → likely to buy)
buy = (0.3 * age + 0.0002 * salary + np.random.normal(0, 1, n)) > 20
buy = buy.astype(int)

# Create DataFrame
df = pd.DataFrame({
    'Age': age,
    'Salary': salary,
    'Buy': buy
})
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Age','Salary']]
X
```

```python
y = df['Buy']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    X, y, train_size=0.8, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model
```

```python
model.fit(xtrain,ytrain)
```

```python
ypredict = model.predict(xtest)
```

```python
import  matplotlib.pyplot as plt
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypredict)
```
