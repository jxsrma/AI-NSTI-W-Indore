```python
import numpy as np
import pandas as pd

# Set a random seed for reproducibility
np.random.seed(42)

# Generate sample data
data = {
    'Age': np.random.randint(18, 65, 1000),  # Ages between 18 and 65
    'AnnualIncome': np.random.randint(25000, 120000, 1000),  # Income in INR
    'SpendingScore': np.random.randint(1, 100, 1000),  # Spending score (1 to 100)
    'Tenure': np.random.randint(1, 15, 1000),  # Years of being a customer
    'PurchaseAmount': np.random.uniform(1000, 20000, 1000)  # Purchase amount in INR
}

# Create a DataFrame
df = pd.DataFrame(data)

df.head()
```

```python
df
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Standard Scaler
ss = StandardScaler()

# Standard Scaler values
standardValues = ss.fit_transform(df)

# PCA Model
pcaModel = PCA(n_components=2)

# Normal data PCA
reducedDataNormal = pcaModel.fit_transform(df)

# Standard data PCA
reducedDataSS = pcaModel.fit_transform(standardValues)

print(reducedDataNormal)
print(reducedDataSS)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(reducedDataNormal[:,0],reducedDataNormal[:,1])
```

```python
plt.scatter(reducedDataSS[:,0],reducedDataSS[:,1])
```
