```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('synthetic_student_health_data.csv')
df
```

```python

le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df
```

```python
# df = df[['age',	'gender',	'sleep_quality',	'meal_quality',	'exercise_frequency',	'hours_studied',	'part_time_work_hours',
#          'mental_health_score',	'exam_stress_level',	'gpa',	'passed_gpa',	'healthy_lifestyle',	'cluster_type']]
df.drop('student_id',inplace=True,axis=1)
df
```


## ðŸ“ˆ Regression Models
 - Build a Linear Regression model to predict gpa using all other numeric features.
 - Evaluate the model using MAE, RMSE, and RÂ² score.
 - Visualize predicted vs actual GPA with a scatter plot.
 - Compare prediction results for mental_health_score using a regression model â€” which features are more predictive?

```python
X = df.drop('gpa',axis=1)
X
```

```python
y = df['gpa']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
print("MAE: ", mean_absolute_error(y_test, y_pred))
print("RMSE: ", root_mean_squared_error(y_test, y_pred))
print("R2: ", r2_score(y_test, y_pred)*100)
```


## ðŸ§© Clustering
 - Cluster students into 3 groups using KMeans on sleep_quality, exam_stress_level, and mental_health_score.
 - Reduce dimensionality with t-SNE and visualize clusters with color-coded points.


```python
features = df[['sleep_quality','exam_stress_level','mental_health_score']]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
reduced_Features = pca.fit_transform(features)
plt.scatter(reduced_Features[:,0],reduced_Features[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
KMModel = KMeans(n_clusters=3)
KMModel.fit(reduced_Features)
```

```python
df['cluster'] = KMModel.predict(reduced_Features)
```

```python
from sklearn.manifold import TSNE
```

```python
TSNEModel = TSNE(n_components=2)
visual = TSNEModel.fit_transform(df)
visual = pd.DataFrame(visual,columns=['a','b'])
visual = pd.concat([visual,df['cluster']],axis=1)
visual
for i in sorted(visual['cluster'].unique()):
    data = visual[visual['cluster'] == i]
    plt.scatter(data.a, data.b)
plt.show()
```


## âš™ Support Vector Machine (SVM)
 - Use SVM to classify whether a student passed_gpa using all features.
 - Try both linear and RBF kernels and compare their performance metrics.

```python
from sklearn.svm import SVC
```

```python
X = df.drop('passed_gpa',axis=1)
y = df['passed_gpa'].apply(lambda x: round(x))
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
SVCLModel = SVC(kernel='linear')
SVCLModel.fit(x_train,y_train)
y_predSVL = SVCLModel.predict(x_test)

SVCRModel = SVC(kernel='rbf')
SVCRModel.fit(x_train,y_train)
y_predSVR = SVCRModel.predict(x_test)
```

```python
print("F1 Score SVM with Linear: ",f1_score(y_test,y_predSVL)*100)
print("F1 Score SVM with RBF: ",f1_score(y_test,y_predSVR)*100)
```
