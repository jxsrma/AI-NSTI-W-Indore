```python
dataset = "D:\\EduNet\\NSTI Indore\\AI\\Test\\DL\\Churn Data Customers.csv"
```

```python
import pandas as pd
```

```python
df = pd.read_csv(dataset)
df
```

```python
df.drop('customerID',axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
df
```

Splitting the Data into Features and Labels

X = All Coloumns excluding Churn
<br>
y = Churn

```python
X = df.drop('Churn',axis=1)
X
```

```python
y = df['Churn']
y
```

Standard Scaling of Featurs i.e.: X

```python
ss = StandardScaler()
x_scaled = ss.fit_transform(X)
x_scaled
```

Splitting the data into Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y, train_size= 0.8,random_state=42)
```

Now We will Import the following:
- Tensorflow
- Squential
- Dense

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Single Layer Perceptron:

```python
x_train.shape
```

```python
slp = Sequential([
    
    Dense(1,activation='sigmoid',input_shape=(x_train.shape[1],))
    
])
```

Compilation of the SLP using Adam and Binary Cross-Entropy

```python
slp.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
slp.summary()
```

Training of the Perceptron

```python
slp.fit(x_train, y_train, epochs=40,batch_size=64,validation_data=(x_test,y_test))
```

Now for Multilayer Perceptron

```python
mlp = Sequential([
    Dense(64, activation='tanh', input_shape=(x_train.shape[1],)),
    Dense(32, activation='tanh'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

Compilation of the MLP using Adam and Binary Cross-Entropy

```python
mlp.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
mlp.summary()
```

Training of the Layers

```python
mlp.fit(x_train, y_train, epochs=40,batch_size=64,validation_data=(x_test,y_test))
```
