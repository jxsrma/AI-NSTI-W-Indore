# Image Transformations

```python
# Load image
image = cv2.imread('saved_aimdBanner.jpg')
image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
```

```python
# Resize image
# resized_image = cv2.resize(image, (0,0), fx=0.5, fy=0.5) #(50% smaller)

# Change the size to 500 pixels wide and 300 pixels tall
resized_image = cv2.resize(image, (500, 300))  # Resize it

plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)), plt.title('Resized') # Convert to RGB and display the image
```

```python
print(image.shape, image.shape[:2])
```

```python
# Rotate image
height, width = image.shape[:2]

# Find the center of the picture
center = (width // 2, height // 2)  # Calculate center point
print(center)

# Make a plan to rotate 45 degrees
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)

# Rotate the picture
rotated = cv2.warpAffine(image, rotation_matrix, (width, height)) # Apply rotation

plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)), plt.title('Rotated')
```

```python
# Flip image (mirror effect)

# Flip left to right
# flipped_horizontal = cv2.flip(image, 1)  # Horizontal flip
# plt.imshow(cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB)), plt.title('Flipped')

# Flip up to down
flipped_vertical = cv2.flip(image, 0)  # Vertical flip
plt.imshow(cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB)), plt.title('Flipped')

# Flip both ways
# flipped_both = cv2.flip(image, -1)  # Both flips
# plt.imshow(cv2.cvtColor(flipped_both, cv2.COLOR_BGR2RGB)), plt.title('Flipped')
```

```python
# Crop image (focus on face)

# Cut out a rectangle (rows 50 to 250, columns 100 to 400)
cropped = image[50:250, 100:400]  # Crop the region

plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)), plt.title('Cropped')
```

```python
# Display results
plt.figure(figsize=(12,8))
plt.subplot(231), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)), plt.title('Resized')
plt.subplot(233), plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)), plt.title('Rotated')
plt.subplot(234), plt.imshow(cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB)), plt.title('Flipped')
plt.subplot(235), plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)), plt.title('Cropped')
plt.show()
```

# Pixel Operations with NumPy

```python
# Load image
image = cv2.imread('saved_aimdBanner.jpg')

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
```

```python
# Show a tiny part of the picture (first 5x5 pixels)
print("First 5x5 pixels of the picture:\n", image[0:5, 0:5])  # Print small array
```

```python
# Access pixel values
print("Pixel at (100,50):", image[100, 50])  # [B, G, R]  Blue, Green, Red
```

```python
# Make a copy of the picture to change
modified_image = image.copy()  # Copy the image
```

```python
# Change one pixel to blue at (100, 50)
modified_image[100, 50] = [255, 0, 255]  # Set to Blue (BGR)

plt.imshow(cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB))
plt.title('Modified Pixel Art')
plt.show()
```

```python
# Change pixel colors

# Make top-left corner red
modified_image[0:50, 0:50] = [0, 0, 255]  # BGR: [0,0,255] = 🔴

plt.imshow(cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB))
plt.title('Modified Pixel Art')
plt.show()
```

```python
# Create white rectangle
# cv2.rectangle(modified_image, (100,100), (200,200), (255,0,0), -1)

# Change a rectangle to white (rows 50-100, columns 100-200)
modified_image[50:100, 100:200] = [255, 0, 255]  # Set to pink

plt.imshow(cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB))
plt.title('Modified Pixel Art')
plt.show()
```

```python
# Draw green text
cv2.putText(image, 'AIMD', (150,300), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,255,0), 3)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Modified Pixel Art')
plt.show()
```
