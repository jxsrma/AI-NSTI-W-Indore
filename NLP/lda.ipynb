```python
documents = [
    "I love playing football with my friends on weekends.",
    "The government passed a new law on environmental protection.",
    "Artificial Intelligence is transforming the future of technology.",
    "I enjoy baking cakes and cookies during the holidays.",
    "The stock market crashed due to inflation and rising interest rates.",
    "We went hiking in the mountains and saw beautiful landscapes.",
    "Machine learning models require a lot of data to train effectively.",
    "The president gave a speech about the economic policies.",
    "Reading mystery novels is one of my favorite hobbies.",
    "Deep learning is a subset of machine learning in data science."
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
bow = CountVectorizer(stop_words='english')
X = bow.fit_transform(documents)
X.toarray()
```

```python
bow.get_feature_names_out()
```

```python

lda = LatentDirichletAllocation(n_components=3,random_state=42)
lda.fit(X)

text = bow.get_feature_names_out()
for idx, topic in enumerate(lda.components_):
    print(idx+1,[text[i] for i in topic.argsort()[:-5- 1:-1]])
    print(topic.argsort()[:-3 - 1:-1])
    # print(f"\nTopic {idx + 1}:")
    # print(", ".join([text[i] for i in topic.argsort()[:-3 - 1:-1]]))
```
